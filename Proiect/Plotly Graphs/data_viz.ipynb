```python
country_data = pd.read_csv("https://raw.githubusercontent.com/Lys4nder/SchoolDataset/main/data_project_full.csv",sep=';',
                          dtype={"pop": int})
country_data.head()
```

```python
map_fig = px.scatter_geo(country_data, locations ='iso_alpha',
                         projection = "natural earth",
                         size='pop',
                         hover_name='country',
                         color='pop',
                         title='Țările Artiștilor'
                        )
map_fig.show()
#map_fig.write_html("/Users/lysanderpitu/desktop/map.html")
```

```python
fig = px.line(x=["1","2","3","4","5"], 
              y=[5201316, 4935093, 4835519, 5619357, 4785117], title="Venituri medii", labels={'x':"Saptamana", 'y':'Bani generati $'} )
fig.show()
```

```python
from plotly.subplots import make_subplots
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(x=["1","2","3","4","5"], y=[5201316, 4935093, 4835519, 5619357, 4785117], name="Medii"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=["1","2","3","4","5"], y=[950770423, 987018583, 967103798,  1123871497, 957023388], name="Totale"),
    secondary_y=True,
)

fig.update_layout(
    title_text="Comparare Venituri"
)

fig.update_xaxes(title_text="Saptamani")
fig.update_yaxes(title_text="<b>Medii</b>", secondary_y=False, color="blue")
fig.update_yaxes(title_text="<b>Totale</b>", secondary_y=True, color="red")


fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=["1","2","3","4","5"], y=[5201316, 4935093, 4835519, 5619357, 4785117], name="Medii"),
)


fig.add_trace(go.Scatter(
    x=["1","2","3","4","5"], y=[950770423, 987018583, 967103798,  1123871497, 957023388], name="Totale",
    yaxis="y2"
))

fig.add_trace(go.Scatter(
    x=["1","2","3","4","5"],
    y=[34382271,  30692698,  31206541, 32853130,  27942865],
    name="Maxim Saptamanal",
    yaxis="y3"
))

fig.update_layout(
    xaxis=dict(
        domain=[0.3, 0.7]
    ),
    yaxis=dict(
        title="Medii",
        titlefont=dict(
        ),
        tickfont=dict(
        )
    ),
    yaxis2=dict(
        title="Totale",
        titlefont=dict(
        ),
        tickfont=dict(
        ),
        anchor="free",
        overlaying="y",
        side="left",
        position=0.15
    ),
    yaxis3=dict(
        title="Maxime",
        titlefont=dict(
            color="#FF0000"
        ),
        tickfont=dict(
            color="#FF0000"
        ),
        anchor="x",
        overlaying="y",
        side="right"
    ),
)
#de lucrat la culori mai sus
fig.update_layout(
    title_text="Comparare Venituri Medii, Totale, Maxime Saptamanale",
    width=1000,
)

fig.show()
```

```python
import pandas as pd
import numpy as np
import cufflinks as cf
import plotly as py
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline

from plotly.offline import download_plotlyjs, init_notebook_mode,plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
song_data = pd.read_csv("https://raw.githubusercontent.com/Lys4nder/SchoolDataset/main/song_data.csv",sep=';')
song_data.tail()
```

```python
fig = px.pie(song_data, names="colab", color='colab',
                labels={'colab':'Colaborare'},
                 title="Procentaj Colaborari",
            color_discrete_map={'Da':'cyan',
                                 'Nu':'blue'})
fig.show()
```

```python
fig = px.pie(song_data, names="gen", title="Genuri")
fig.show()
```

```python
fig=px.scatter(song_data, x="vizualizari", y="venituri", hover_name = "nume", color="artist", log_x=True, size_max=60
               )
fig.show()
#size="vizualizari"
```

```python
fig = px.histogram(song_data, x="lansare", nbins=47)
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(song_data, x="artist", title="Câte melodii a scos fiecare artist?")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(song_data, x="nume", y="vizualizari", title="Vizualizari/Melodie")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(song_data, x="artist", y="vizualizari", title="Vizualizări individuale/artist")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.scatter_3d(song_data, x="vizualizari", y="durata", color="artist", z='venituri', hover_name="nume")
fig.show()
```
